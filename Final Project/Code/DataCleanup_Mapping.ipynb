## Mapping nearest census DAs

```python
#Read DA Excel file

path="../Data/DA centre lat long/DA_centre_lat_long_Ontario_cleaned.xlsx"

da_df=pd.read_excel(path)
da_df.head()
```

```python
#use scipy.spatial.KDTree.query to find nearest neighbor
from scipy import spatial
tree = spatial.KDTree(list(zip(da_df['DA_lat'], da_df['DA_long'])))

tree.data
```

```python
school_loc = list(zip(schools_eqao['Latitude'],schools_eqao['Longitude']))

n=len(school_loc)
```

```python
# da_df.loc[8830]['DA_id']

Nearest4DA=[]

for i in range (0,n) :
    Nearest4DA.append(list(da_df.loc[tree.query(school_loc[i],k=4, eps=0, p=2)[1]]['DA_id']))
    
schools_eqao['Nearest4DA']=Nearest4DA
```

```python
#creating bins
bins = [0, 60, 80, 100]
labels=['Bad','Fair','Excellent']
df_ml['EQAO_Class']= pd.cut(df_ml['Avg_EQAO'],bins,labels=labels)
df_ml.head()
```

```python
#Start Model building
#Create Target and Predictors and split the data in test and training

x = df_ml.drop(['EQAO_Class','Avg_EQAO'],axis=1)
y = df_ml['EQAO_Class']
```

```python
# Models in scikitlearn can only handle numeric variables so use one-hot-encoding to create dummy variables
enc = OneHotEncoder(handle_unknown='ignore')
enc.fit(x)
x_ohe = enc.transform(x)
```

```python
# Split the data into test and training with 15% reserved for testing
x_train,x_test,y_train,y_test = train_test_split(x_ohe,y, test_size = 0.15)
```

```python
y_test.value_counts()
```

## XGBoost

```python
# Compute class weights to use in balancing the data
class_weights = class_weight.compute_class_weight('balanced',
                                                 np.unique(y_train),
                                                 y_train)
print(class_weights)
```

```python
# Set the evaluation criteria for the model. 
eval_set = [(x_train, y_train), (x_test, y_test)]
eval_metric = ["logloss","error"]
```

```python
# define the model parameters. this is a starting point, we can refine these later using grid search.
param = {}
param['objective'] = 'multi:softmax'
param['eta'] = 0.1 #learningrate
param['ntrees'] = 300 
param['subsample'] = 0.7
param['silent'] = 0
param['n_jobs'] = 4
param['max_depth'] = 6
param['colsample_bytree'] = 0.6
param['gamma'] = 0.1
param['reg_lambda'] = 0.1
param['min_child_weight']= 3
# param['scale_pos_weight']= 0.15
```

```python
xgb_model =  xgb.XGBClassifier(**param,random_state=87,eval_metric = eval_metric,eval_set = eval_set,verbosity=3\
                               ,class_weight=class_weight)
```

```python
# Fit the model on training data and cross validate 
xgb_model.fit(x_train,y_train)
# scores = cross_val_score(xgb_model, x_test,y_test, cv=5,scoring='f1_macro')
# scores 
```

```python
y_pred = xgb_model.predict(x_test)
```
